```python

import numpy as np
import os

# --- PASO 1: CONFIGURACIÓN DE RUTAS Y NOMBRES ---
# Define tus 4 archivos con sus nombres específicos
# AJUSTA los nombres de archivo y las rutas según corresponda.

# IMPORTANTE: Usamos 'r' al inicio para evitar el error de unicode escape.
archivos_info = {
    'clientes': r'C:\Users\JUANK\Documents\Curso de IBM\Aurelion Grupo\GUAYERD-IA\datos\clientes.xlsx',
    'productos': r'C:\Users\JUANK\Documents\Curso de IBM\Aurelion Grupo\GUAYERD-IA\datos\productos.xlsx',
    'ventas': r'C:\Users\JUANK\Documents\Curso de IBM\Aurelion Grupo\GUAYERD-IA\datos\ventas.xlsx',
    'detalle_ventas': r'C:\Users\JUANK\Documents\Curso de IBM\Aurelion Grupo\GUAYERD-IA\datos\detalle_ventas.xlsx'
}

dataframes = {}

print("--- 1. CARGANDO DATASETS INDIVIDUALMENTE ---")

for nombre, ruta in archivos_info.items():
    try:
        # Usamos pd.read_excel para leer cada archivo
        df_temp = pd.read_excel(ruta)
        dataframes[nombre] = df_temp  # Almacena cada DataFrame con su nombre clave
        print(f"✅ Cargado: {nombre} ({df_temp.shape[0]} filas)")
    except FileNotFoundError:
        print(f"❌ ERROR: Archivo '{nombre}' no encontrado en la ruta: {ruta}")
    except Exception as e:
        print(f"❌ ERROR al procesar '{nombre}': {e}")

# Accede a tus DataFrames así:
# df_clientes = dataframes['clientes']
# df_productos = dataframes['productos']
# ...

# ----------------------------------------------------------------------
# --- PASO 2: FUNCIÓN DE LIMPIEZA Y EDA INDIVIDUAL ---
# ----------------------------------------------------------------------

def analizar_dataframe(df_nombre, df):
    """Realiza las funciones principales de limpieza y EDA en un DataFrame."""
    print("\n" + "="*70)
    print(f"--- INICIANDO ANÁLISIS PARA: {df_nombre.upper()} ---")
    print(f"Filas iniciales: {df.shape[0]}, Columnas: {df.shape[1]}")
    print("="*70)

    # 1. Búsqueda y Eliminación de Duplicados
    conteo_duplicados = df.duplicated().sum()
    print(f"1. Duplicados: {conteo_duplicados} filas duplicadas encontradas.")
    
    if conteo_duplicados > 0:
        df_limpio = df.drop_duplicates(ignore_index=True)
        print(f"   > Filas eliminadas: {conteo_duplicados}. Filas restantes: {df_limpio.shape[0]}")
    else:
        df_limpio = df.copy()
        print("   > No se eliminaron duplicados.")
        
    print("-" * 35)

    # 2. Conteo de Nulos
    nulos_info = pd.DataFrame({
        'Total Nulos': df_limpio.isnull().sum(),
        'Porcentaje Nulos': (df_limpio.isnull().sum() / len(df_limpio)) * 100
    })
    nulos_info_filtrada = nulos_info[nulos_info['Total Nulos'] > 0].sort_values(
        by='Total Nulos', ascending=False
    )
    
    print("2. Conteo de Valores Nulos:")
    if nulos_info_filtrada.empty:
        print("   > ✅ No se encontraron valores nulos.")
    else:
        print(nulos_info_filtrada)
        
    print("-" * 35)

    # 3. Conteo de Valores Únicos por Columna
    print("3. Conteo y Muestra de Valores Únicos:")
    for col in df_limpio.columns:
        n_unicos = df_limpio[col].nunique()
        
        print(f"   > '{col}' | Únicos: {n_unicos} | Tipo: {df_limpio[col].dtype}")
        
        # Muestra valores si hay 50 o menos
        if n_unicos <= 50:
            valores = df_limpio[col].unique().tolist()
            if len(valores) > 10:
                 print(f"     Primeros 10 valores: {valores[:10]}...")
            else:
                 print(f"     Valores: {valores}")
        else:
            print("     Valores omitidos (más de 50 únicos).")
            
    return df_limpio # Retorna el DataFrame limpio

# --- PASO 3: EJECUTAR EL ANÁLISIS PARA CADA DATAFRAME CARGADO ---

dataframes_limpios = {}

for nombre, df in dataframes.items():
    df_limpio = analizar_dataframe(nombre, df)
    dataframes_limpios[f'df_{nombre}_limpio'] = df_limpio

# --- FINALIZACIÓN ---
print("\n" + "="*70)
print("ANÁLISIS COMPLETO.")
print("Ahora tienes los DataFrames limpios accesibles por nombre:")
for key in dataframes_limpios.keys():
    print(f"- {key}")
```
